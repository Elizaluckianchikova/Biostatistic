```python
# Функция считающая частоты:
def cat_describe(column):
    count = column.count()
    abs_freq = column.value_counts()
    rel_freq = (abs_freq*100/count)

    count = pd.Series(count, index=['Count'])
    abs_freq = abs_freq.add_suffix('_abs_freq')
    rel_freq = rel_freq.add_suffix('_rel_freq')
    column_stat = pd.concat([count, abs_freq, rel_freq])
    return column_stat
```

```python
df_cat.apply(cat_describe,axis=0)
```

```python
df_cat_stats = df_cat.apply(cat_describe,axis=0)
new_order_index = ['Count'] + df_cat_stats.index.drop('Count').to_list()
df_cat_stats = df_cat_stats.loc[new_order_index,]
df_cat_stats
```

**Задание 2**

Подготовьте программу, которая будет формировать рандомизационный список для клинического исследования.

Представьте, что вы проводите исследование, в которое нужно включить 64 добровольца. Каждый доброволец (иначе – субъект) должен получить терапию, обозначенную A, или же другую терапию, обозначенную B. Желательно применить блоковую рандомизацию.

Итоговый рандомизационный список должен содержать

порядок рандомизации (от 1 до 64),
код терапии (вектор значений A или B длиной 64),
рандомизационный номер (числа от 1 до 64, совпадает с порядком рандомизации; чтобы он отличался от порядка рандомизации, можете добавить к нему какой-нибудь произвольный префикс).
Список нужно выгрузить в файл в формате .csv или .xlsx.

```python
from numpy.random import choice
import numpy as np
```

```python
sample_size = 64
n_batches = 16
groups = ['A','B']
```

```python
blocks_ind = np.arange(1,sample_size+1).reshape(n_batches,int(sample_size/n_batches))
blocks_group = np.array([choice(groups, len(i)) for i in blocks_ind])
```

```python
df_randomized = pd.DataFrame({'Randomize_order' : blocks_ind.flatten(),
                              'Therapy_code' : blocks_group.flatten()})

df_randomized['Randomize_number'] = 'NUM' + df_randomized['Randomize_order'].astype('str')
df_randomized.head(15)
```

```python
df_randomized.to_excel('Block_randomized_64.xlsx',index=False)
```
