```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import torch
from torch import nn
from torch import optim
from torch.utils.data import Dataset, DataLoader
```

```python
X, y = islanders[['age', 'Happy_Sad_group', 'Dosage', 'Drug', 'Mem_Score_Before']], islanders['Diff']
```

```python
class Data(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X.astype(np.float32))
        self.y = torch.from_numpy(y.astype(np.float32))
        self.len = self.X.shape[0]

    def __getitem__(self, index):
        return self.X[index], self.y[index]

    def __len__(self):
        return self.len
```

```python
batch_size = 200

X_train, X_test, y_train, y_test = train_test_split(X.to_numpy(), y.to_numpy(), test_size=.25, random_state=26, shuffle=True)
```

```python
# Instantiate training and test data
train_data = Data(X_train, y_train)
train_dataloader = DataLoader(dataset=train_data, batch_size=batch_size, shuffle=True)

test_data = Data(X_test, y_test)
test_dataloader = DataLoader(dataset=test_data, batch_size=batch_size, shuffle=True)

# Check it's working
for batch, (X, y) in enumerate(train_dataloader):
    print(f"Batch: {batch+1}")
    print(f"X shape: {X.shape}")
    print(f"y shape: {y.shape}")
    break
```

```python
class SimpleNet(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(SimpleNet, self).__init__()
        self.hidden_dim = hidden_dim
        self.drug_embed = nn.Embedding(3, hidden_dim)
        self.hs_group_embed = nn.Embedding(2, hidden_dim)
        self.dosage_embed = nn.Embedding(4, hidden_dim)

        hidden_dim = hidden_dim * input_dim
        self.layers_conv = nn.Sequential(
            nn.Conv1d(hidden_dim, hidden_dim, 1),
            nn.Dropout(),
            nn.Tanh(),
            nn.Conv1d(hidden_dim, hidden_dim, 1),
            nn.Dropout(),
            nn.Tanh(),
            nn.Conv1d(hidden_dim, hidden_dim, 1),
            nn.Tanh(),
        )
        self.fcn = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.Linear(hidden_dim, output_dim)
        )

    def forward(self, x):
        x1 = x[:, 0].unsqueeze(-1).repeat(1, self.hidden_dim)
        x2 = self.hs_group_embed(x[:, 1].long()).squeeze(-1)
        x3 = self.dosage_embed(x[:, 2].long()).squeeze(-1)
        x4 = self.drug_embed(x[:, 3].long()).squeeze(-1)
        x5 = x[:, 4].unsqueeze(-1).repeat(1, self.hidden_dim)
        x = torch.cat([x1, x2, x3, x4, x5], axis=1).unsqueeze(-1)
        x = self.layers_conv(x)
        return self.fcn(x.squeeze(-1))
```

```python
input_dim = 5
hidden_dim = 10
output_dim = 1

model = SimpleNet(input_dim, hidden_dim, output_dim)
learning_rate = 3e-5
loss_fn = nn.MSELoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate, weight_decay=0.1)

model
```

```python
model.train()
num_epochs = 20000
loss_values = []

for epoch in range(num_epochs):
    for X, y in train_dataloader:
        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        pred = model(X)
        loss = torch.sqrt(loss_fn(pred, y.unsqueeze(-1)))
        loss_values.append(loss.item())
        loss.backward()
        optimizer.step()
```

```python
plt.plot(loss_values)
```

```python
model.eval()
with torch.no_grad():
        for X, y in test_dataloader:
                pred = model(X)
                loss = torch.sqrt(loss_fn(pred, y.unsqueeze(-1)))
                print(loss)
                print(r2_score(y.numpy(), pred.numpy()))
```
